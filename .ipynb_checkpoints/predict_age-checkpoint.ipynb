# Problem Statement
The task is to predict the age of a person from his or her facial attributes. For simplicity, the problem has been converted to a multi-class problem with classes as Young, Middle and Old.


## Dataset
Indian Movie Face database (IMFDB) is a large unconstrained face database consisting of 34512 images of 100 Indian actors collected from more than 100 videos. All the images are manually selected and cropped from the video frames resulting in a high degree of variability interms of scale, pose, expression, illumination, age, resolution, occlusion, and makeup. IMFDB is the first face database that provides a detailed annotation of every image in terms of age, pose, gender, expression and type of occlusion that may help other face related applications. For information about is IMFDB is available at http://cvit.iiit.ac.in/projects/IMFDB/.

The dataset is cleaned and formatted to give total of 26742 images with 19906 images in train and 6636 images in test.

The task is to predict the age of a person from his or her facial attributes. For simplicity, the problem has been converted to a multiclass problem with classes as Young, Middle and Old.



```python
% matplotlib inline
# import modules
import os
import random

import pandas as pd
from scipy.misc import imread
```

```python
# Read train and test data files
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

Lets focus on one problem: to handle variations in shape.

We will resize all the train and test images to same size and store them in numpy arrays.

```python
# Resize train images
from scipy.misc import imresize

temp = []
for img_name in train.ID:
    img_path = os.path.join('data/Train', img_name)
    img = imread(img_path)
    img = imresize(img, (32, 32))
    img = img.astype('float32') # this will help us in later stage
    temp.append(img)

train_x = np.stack(temp)
```

```python
# Resize test images
temp = []
for img_name in test.ID:
    img_path = os.path.join('data/Test', img_name)
    img = imread(img_path)
    img = imresize(img, (32, 32))
    temp.append(img.astype('float32'))

test_x = np.stack(temp)
```

We can do one more thing that could help us build a better model; i.e. we can normalize our images. Normalizing the images will make our train faster.

```python
train_x = train_x / 255.
test_x = test_x / 255.
```

Now lets take a look at distribution of our target variable.

```python
train.Class.value_counts(normalize=True)
```

On the basis of distribution of our data, we can make a simple prediction as most of the actors are MIDDLE aged.

### Let's solve the problem

Before creating something substantial, let us bring our target variable in shape. We will convert our target into dummy columns so that it will be easier for our model to ingest it.

```python
import keras
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
train_y = lb.fit_transform(train.Class)
train_y = keras.utils.np_utils.to_categorical(train_y)
```

Now comes the main part, building a model! As the problem is related to image processing, it is wiser to use neural networks to solve the problem. 
Lets first try a simple feedforward neural network.

But before that we should specify all the parameters we will be using in our network

```python
input_shape = (32, 32, 3)
hidden_units = 500
output_classes = 3

epochs = 5
batch_size = 128
```

```python
# Import necessary keras modules
from keras.models import Sequential
from keras.layers import Dense, Flatten, InputLayer, Conv2D

# Build a model with above specified parameters
model = Sequential([
  InputLayer(input_shape=input_shape),
  Flatten(),
  Dense(units=hidden_units, activation='relu'),
  Dense(units=output_classes, activation='softmax'),
])

# To see how our model looks like; lets print it
model.summary()
```

Now lets compile our model and train it.

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_x, train_y, batch_size=batch_size,epochs=epochs,verbose=1)
```

We have trained the model but, we still haven’t validated it
Now we will validate our model to ensure that our model will perform well on both the data it is training on and on a new testing data.
We will use 20% of the train data as validation data.

```python
model.fit(train_x, train_y, batch_size=batch_size,epochs=epochs,verbose=1, validation_split=0.2)
```

So our model gives about 63% of accuracy on validation data, which is better than predicting all the actors as middle aged(nearly 54%). 

#### Visual Inspection of our model
Lets print the image along with the predictions of our trained model.

```python
i = random.choice(train.index)
img_name = train.ID[i]

img = imread(os.path.join('data/Train', img_name)).astype('float32')
imshow(imresize(img, (128, 128)))
pred = model.predict_classes(train_x)
print('\nOriginal:', train.Class[i], 'Predicted:', lb.inverse_transform(pred[i]))
```
